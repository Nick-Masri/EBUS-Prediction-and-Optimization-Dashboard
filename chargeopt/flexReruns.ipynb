```python
import pandas as pd
```

```python

df = pd.read_csv('results.csv')

df.head()
```

```python
df = df[df['case_name'].str.contains('flex')]
df.head()
```

```python
df.routes.unique()
```

```python
dynamic_routes = set()
rows_to_keep = []

for index, row in df.iterrows():
    if row['type'] == 'dynamic':
        dynamic_routes.add(row['routes'])
    elif row['type'] == 'static' and row['routes'] in dynamic_routes:
        dynamic_routes.remove(row['routes'])

for index, row in df.iterrows():
    if row['type'] == 'dynamic' and row['routes'] in dynamic_routes:
        rows_to_keep.append(row)

df_result = pd.DataFrame(rows_to_keep)
print(df_result)
```

```python
import yaml
import subprocess
import ast

# for all of the above entries, run them statically
for idx, row in df_result.iterrows():

    #############
    # first run
    #############
    # Load the initial configuration
    with open('config.yml', 'r') as f:
        config = yaml.safe_load(f)

    # use the same name and route params
    config['filename'] = row['case_name']
    config['routes'] = ast.literal_eval(row['routes'])
    # # run it this time as static
    config['runType'] = 'static'

    # Save the modified configuration
    with open('config.yml', 'w') as f:
        yaml.safe_dump(config, f)

    # Call main.py
    subprocess.call(['python3', 'main.py'])
```
