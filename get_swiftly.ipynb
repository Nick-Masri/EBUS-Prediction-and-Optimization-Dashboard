```python
n = len(vehicle_pos.json()['entity'])
coaches = ['7501', '7502', '7503', '7504', '7505', '9501', '9502', '9503', '9504', '9505']
active_coaches = []
for i in range(n):
    if vehicle_pos.json()['entity'][i]['vehicle']['vehicle']['id'] in coaches:
        print(vehicle_pos.json()['entity'][i]['vehicle'])
        active_coaches.append(vehicle_pos.json()['entity'][i]['vehicle']['vehicle']['id'] )
```

```python
import datetime

def convert_to_pst(seconds):
    # The number of seconds since January 1, 1970 00:00:00 UTC
    # seconds = int(rt_trip.json()['entity'][i]['tripUpdate']['stopTimeUpdate'][-1]['arrival']['time'])

    # Create a datetime object from the number of seconds
    dt = datetime.datetime.utcfromtimestamp(seconds)

    # Convert the datetime object to the Pacific Standard Time (PST) time zone
    dt_pst = dt.astimezone(datetime.timezone(datetime.timedelta(hours=-8)))

    # Format the date and time in the h:mm AM/PM format
    time_str = dt_pst.strftime("%I:%M %p")

    # Print the result
    return time_str
```

```python
n = len(rt_trip.json()['entity'])

arrival_time = {}
for i in range(n):
    # if rt_trip.json()['entity'][i]['vehicle']['vehicle']['label'] in veh_id:
    if len(rt_trip.json()['entity'][i]['tripUpdate'].keys()) > 2:
        coach = rt_trip.json()['entity'][i]['tripUpdate']['vehicle']['id']
        if coach in active_coaches:
            pst = convert_to_pst(int(rt_trip.json()['entity'][i]['tripUpdate']['stopTimeUpdate'][-1]['arrival']['time']))
            print(pst)
            arrival_time[coach] = pst
        # print(rt_trip.json()['entity'][i]['tripUpdate']['stopTimeUpdate'][-1])
        # break
```
